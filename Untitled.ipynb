# College Tour Information Scraper

Gathers information from the https://www.youvisit.com/collegesearch/ website



```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
s = requests.Session()
r = s.get("https://www.youvisit.com/tour/60281", headers = {'User-Agent': 'Mozilla/5.0'})
r = s.get("https://www.youvisit.com/tour/ada/60281/80230" , headers = {'User-Agent': 'Mozilla/5.0'})
soup = bs(r.text, 'html.parser')
print(soup.prettify())
```
